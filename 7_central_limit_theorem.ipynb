## Notes on Central Limit Theorem (CLT)

---

### 1. What is the CLT?
If you take many random samples from any population and calculate their means, the distribution of those means will resemble a normal (bell-shaped) curve—even if the original data is not normal.

---

### 2. When Does CLT Work Best?
- **Large sample size:** Usually, n > 30 is considered sufficient.
- **Random and independent samples:** Each sample should be chosen randomly and independently from the population.

---

### 3. Why is CLT Important?
- **Predictability:** Allows us to use normal distribution formulas to make predictions about sample means, regardless of the population's original distribution.
- **Real-world relevance:** Explains why averages (like test scores, heights, etc.) often follow a bell curve.

---

### 4. Summary
- Take many random samples  
- Calculate their means  
- Plot those means  
- **Result:** The distribution of sample means will be approximately normal!

---

> **In essence:** The CLT is a foundational concept in statistics that enables powerful and practical analysis, even when the underlying data isn't normally distributed.
